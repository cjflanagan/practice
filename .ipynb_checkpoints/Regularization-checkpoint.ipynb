## Dealing with an "ill-posed"/"undetermined" dataset 
If a dataset has significantly more dimensions than observations it is still possible to fit a model but it must be done with care as it is very easy to over-fit as there can be infinitly many solutions to techniques such as standard least squares regression. 

Ways of dealing with such a problem are to impose contraints. Common techniques for regression problems are lasso and ridge regression 


```python
from sklearn.datasets import load_boston
boston = load_boston()
print(boston['data'])
```

```python
import pandas as pd

boston_df = pd.DataFrame(boston['data'])
boston_df.columns = boston['feature_names']
boston_df.head()
```

```python
Y = boston['target']
X = boston['data'][0:10]
```

```python
from sklearn import linear_model
from sklearn import datasets

lr = linear_model.LinearRegression()
y = boston.target

lr.fit(x, y)
lr.score(x, y)
```
